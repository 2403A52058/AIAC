Task Description #3 – JavaScript Interactivity
Task: Prompt AI to generate a JS script that validates a simple login
form (non-empty username/password).
Expected Output:
Working on submit JS validation.
Clear error messages if inputs are empty.

Prompt:generate a js script that validates a simple login form (non-empty username/password)

```python
html_content = """
<!DOCTYPE html>
<html>
<head>
<title>Login Form Validation</title>
<style>
  .error { color: red; }
</style>
</head>
<body>

<h2>Login</h2>

<form id="loginForm" onsubmit="return validateForm()">
  <div>
    <label for="username">Username:</label>
    <input type="text" id="username" name="username">
    <span id="usernameError" class="error"></span>
  </div>
  <br>
  <div>
    <label for="password">Password:</label>
    <input type="password" id="password" name="password">
    <span id="passwordError" class="error"></span>
  </div>
  <br>
  <button type="submit">Login</button>
</form>

<script>
function validateForm() {
  let username = document.getElementById("username").value;
  let password = document.getElementById("password").value;
  let usernameError = document.getElementById("usernameError");
  let passwordError = document.getElementById("passwordError");
  let isValid = true;

  usernameError.textContent = ""; // Clear previous error message
  passwordError.textContent = ""; // Clear previous error message

  if (username == "") {
    usernameError.textContent = "Username must be filled out";
    isValid = false;
  }
  if (password == "") {
    passwordError.textContent = "Password must be filled out";
    isValid = false;
  }
  return isValid; // Prevent form submission if not valid
}
</script>

</body>
</html>
"""

from IPython.display import HTML
display(HTML(html_content))
```

EXPLANATION:
1. The variable `html_content` stores an HTML page that includes a simple login form.
2. The page has fields for **username** and **password**, each with an area to show error messages.
3. A small CSS rule (`.error { color: red; }`) styles error messages in red.
4. The form calls the JavaScript function `validateForm()` when the user clicks **Login**.
5. Inside `validateForm()`, the input values for username and password are retrieved.
6. Error messages are cleared before new validation checks run.
7. If either field is empty, an error message appears, and form submission is stopped.
8. `display(HTML(html_content))` displays this interactive HTML page inside a Jupyter Notebook.


Task Description #4 – Python Backend Integration
Task: Ask AI to generate a Flask app that serves the HTML form (Task

Prompt:Generate a Flask app that serves an HTML login form (with username and password). On form submission, if login is successful, display “Welcome, [username]!”; otherwise, show “Invalid username or password.” Include all necessary routes and HTML templates in the code.

```python
from flask import Flask, request, render_template_string

app = Flask(__name__)

# HTML for the login form
LOGIN_FORM_HTML = """
<!DOCTYPE html>
<html>
<head>
    <title>Login</title>
</head>
<body>
    <h2>Login</h2>
    <form method="post" action="/">
        <label for="username">Username:</label><br>
        <input type="text" id="username" name="username"><br>
        <label for="password">Password:</label><br>
        <input type="password" id="password" name="password"><br><br>
        <input type="submit" value="Login">
    </form>
</body>
</html>
"""

# HTML for the welcome page
WELCOME_HTML = """
<!DOCTYPE html>
<html>
<head>
    <title>Welcome</title>
</head>
<body>
    <h2>Welcome, {{ username }}!</h2>
</body>
</html>
"""

# HTML for the invalid login page
INVALID_LOGIN_HTML = """
<!DOCTYPE html>
<html>
<head>
    <title>Invalid Login</title>
</head>
<body>
    <h2>Invalid username or password.</h2>
    <p><a href="/">Try again</a></p>
</body>
</html>
"""

@app.route('/', methods=['GET', 'POST'])
def login():
    if request.method == 'POST':
        username = request.form['username']
        password = request.form['password']

        # Simple validation (replace with your actual authentication logic)
        if username == 'user' and password == 'password':
            return render_template_string(WELCOME_HTML, username=username)
        else:
            return render_template_string(INVALID_LOGIN_HTML)

    return render_template_string(LOGIN_FORM_HTML)

if __name__ == '__main__':
    # To run this in Colab, we need to use a library like ngrok
    # to expose the Flask server running on a local port.
    # Alternatively, you can run this code in a local environment.
    print("To run this Flask app in Colab, you'll need to set up a tunnel using a service like ngrok.")
    print("Alternatively, you can copy this code and run it in a local Python environment.")
    # app.run(debug=True) # Uncomment this line to run locally
```
