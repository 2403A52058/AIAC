Task Description #1 – Linear Search implementation
Task: Write python code for linear_search() function to search a value
in a list and extract it’s index

PROMPT:generate a python code for linear_search() funcion to search a value in a list and extract its index

```python
def linear_search(input_list, value):
    """
    Searches for a value in a list using linear search.

    Args:
        input_list: The list to search within.
        value: The value to search for.

    Returns:
        The index of the value if found, otherwise -1.
    """
    for i in range(len(input_list)):
        if input_list[i] == value:
            return i
    return -1

# Example usage:
my_list = [10, 25, 5, 15, 30]
search_value = 15
index = linear_search(my_list, search_value)

if index != -1:
    print(f"Value {search_value} found at index {index}")
else:
    print(f"Value {search_value} not found in the list")

search_value = 50
index = linear_search(my_list, search_value)

if index != -1:
    print(f"Value {search_value} found at index {index}")
else:
    print(f"Value {search_value} not found in the list")
```

EXPLANATION : This code defines a `linear_search` function that searches for a specific value within a list. It iterates through each element and returns the index if the value is found. If the value is not present in the list after checking all elements, it returns -1. The example usage demonstrates how to call the function and print whether the searched value was found and its index.

Task Description #2 – Sorting Algorithms
Task: Ask AI to implement Bubble Sort and check sorted output

PROMPT:generate a python program to implement bubble sort and check sorted output

```python
def bubble_sort(input_list):
    """
    Sorts a list using bubble sort.

    Args:
        input_list: The list to sort.
    """
    n = len(input_list)
    for i in range(n):
        # Last i elements are already in place
        for j in range(0, n - i - 1):
            # Traverse the list from 0 to n-i-1
            # Swap if the element found is greater than the next element
            if input_list[j] > input_list[j + 1]:
                input_list[j], input_list[j + 1] = input_list[j + 1], input_list[j]

# Example usage:
my_list = [64, 34, 25, 12, 22, 11, 90]

print("Original list:", my_list)

bubble_sort(my_list)

print("Sorted list:", my_list)

# Check if the list is sorted
is_sorted = all(my_list[i] <= my_list[i+1] for i in range(len(my_list) - 1))

if is_sorted:
    print("The list is correctly sorted.")
else:
    print("The list is not correctly sorted.")
```

EXPLANATION:This code implements the bubble sort algorithm to sort a list in ascending order. It repeatedly steps through the list, compares adjacent elements, and swaps them if they are in the wrong order. The process is repeated until the list is sorted. The example usage demonstrates the sorting process and checks if the final list is indeed sorted.

PROMPT:Prompt:generate a python program to solve the below case study using linear optimization:
Consider a chocolate manufacturing company that produces only two types of chocolate i.e. A and B. Both the chocolates require Milk and Choco only. Venkataramana
To manufacture each unit of A and B, the following quantities ar required:
Each unit of A requires 1 unit of Milk and 3 units of Choco
Each unit of B requires 1 unit of Milk and 2 units of Choco
 The company kitchen has a total of 5 units of Milk and 12 units of Choco. On each sale, the company makes a profit of Rs 6 per unit A sold and Rs 5 per unit B sold.
Now, the company wishes to maximize its profit. How many units of A and B shoud it produce respectively?

```python
!pip install pulp

from pulp import LpVariable, LpProblem, LpMaximize, lpSum

# Create the problem
prob = LpProblem("Chocolate Manufacturing", LpMaximize)

# Define the decision variables
A = LpVariable("A", lowBound=0, cat='Integer')  # Number of units of chocolate A
B = LpVariable("B", lowBound=0, cat='Integer')  # Number of units of chocolate B

# Define the objective function (maximize profit)
prob += 6 * A + 5 * B, "Total Profit"

# Define the constraints
prob += 1 * A + 1 * B <= 5, "Milk Constraint"  # Milk constraint
prob += 3 * A + 2 * B <= 12, "Choco Constraint" # Choco constraint

# Solve the problem
prob.solve()

# Print the results
print("Optimal production plan:")
print(f"Units of A to produce: {A.varValue}")
print(f"Units of B to produce: {B.varValue}")
print(f"Maximum profit: {prob.objective.value()}")
```

EXPLANATION: This code solves a chocolate manufacturing optimization problem using the `pulp` library.
It defines decision variables `A` and `B` for the number of units of each chocolate type.
The objective is to maximize profit, which is defined as 6*A + 5*B.
Constraints are set for the available milk (1*A + 1*B <= 5) and choco (3*A + 2*B <= 12).
The problem is then solved using the `solve()` method.
Finally, the optimal production quantities for A and B, and the maximum profit are printed.
This demonstrates how to use linear programming to find the best solution given limited resources.

Task Description #4 – Gradient Descent Optimization
Task: Write python code to find value of x at which the function
f(x)=2X3+4x+5 will be minimum

PROMPT:generate a python code to find value of x at which the function f(x)=2X3+4X+5 will be minimum

```python
import sympy as sp

# Define the variable and the function
x = sp.Symbol('x')
f_x = 2*x**3 + 4*x + 5

# Calculate the derivative of the function
f_prime_x = sp.diff(f_x, x)

# Convert the symbolic expressions to numerical functions
f_prime_numerical = sp.lambdify(x, f_prime_x, 'numpy')

# Gradient Descent Implementation
def gradient_descent(starting_point, learning_rate, num_iterations):
    x_current = starting_point
    for i in range(num_iterations):
        gradient = f_prime_numerical(x_current)
        x_current = x_current - learning_rate * gradient
        # Optional: print progress
        # print(f"Iteration {i+1}: x = {x_current}, gradient = {gradient}")
    return x_current

# Set parameters for gradient descent
starting_point = 0.0 # You can choose a different starting point
learning_rate = 0.01
num_iterations = 1000

# Run gradient descent
minimum_x = gradient_descent(starting_point, learning_rate, num_iterations)

print(f"The value of x at which the function is likely minimum is: {minimum_x}")

# Note: For this specific function (2x^3 + 4x + 5), the derivative 6x^2 + 4 is always positive.
# This means the function is always increasing, and there is no local minimum.
# Gradient descent will move towards negative infinity in this case.
# This code demonstrates the gradient descent process, but the function itself doesn't have a minimum.
```

EXPLANATION:

This code uses the `sympy` library to find the minimum of a function using gradient descent.
It defines the variable `x` and the function `f(x) = 2x³ + 4x + 5`.
The derivative of the function is calculated symbolically.
The derivative is then converted to a numerical function for calculations.
The `gradient_descent` function iteratively updates the value of `x`.
It subtracts the product of the learning rate and the gradient from the current x value.
Finally, it prints the value of x where the function is likely minimum, noting that for this specific function, there is no local minimum as the derivative is always positive.
