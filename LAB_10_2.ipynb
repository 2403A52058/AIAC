```python
def calcFact(n):
  """Calculates the factorial of a non-negative integer n."""
  result = 1
  # The loop should iterate from 1 up to n (inclusive) to calculate the factorial.
  for i in range(1, n + 1):
    result = result * i
  return result

def main():
  """Main function to get user input, calculate and print the factorial."""
  while True:
    try:
      # Get input from the user and convert it to an integer.
      num_str = input("Enter a non-negative integer to calculate its factorial: ")
      num = int(num_str)
      # Check if the number is non-negative.
      if num < 0:
        print("Factorial is not defined for negative numbers. Please enter a non-negative integer.")
      else:
        break # Exit the loop if input is valid
    except ValueError:
      print("Invalid input. Please enter an integer.")

  # Call the calcFact function to get the factorial of num.
  FACT = calcFact(num)
  print("The factorial of", num, "is", FACT)

  t = 10
  # Check if the calculated factorial is greater than 10.
  if FACT > t:
    print("BIG Number")
  else:
    print("small number")

# Execute the main function.
main()
```

EXPLANATION:  
1. Function `calcFact(n) computes the factorial of a non-negative integer `n` by multiplying numbers from 1 to `n`.
2. Function `main() repeatedly prompts the user for a non-negative integer, handling invalid or negative input with appropriate messages.
3. Once valid input is received, it calls `calcFact(num)` to compute the factorial.
4. The factorial result is stored in `FACT` and printed.
5. It then checks if the factorial is greater than 10.
6. Based on that check, it prints `"BIG Number"` or `"small number"`.

```python
def generate_fibonacci_series(n_terms):
  """
  Generates a Fibonacci series up to a specified number of terms.

  Args:
    n_terms: The number of terms to generate in the Fibonacci series.

  Returns:
    A list containing the Fibonacci series.
  """
  # Initialize the first two terms of the Fibonacci series
  fibonacci_series = [0, 1]

  # Handle edge cases for n_terms
  if n_terms <= 0:
    return []
  elif n_terms == 1:
    return [0]
  elif n_terms == 2:
    return fibonacci_series

  # Generate the Fibonacci series for n_terms > 2
  while len(fibonacci_series) < n_terms:
    next_term = fibonacci_series[-1] + fibonacci_series[-2]
    fibonacci_series.append(next_term)

  return fibonacci_series

def main():
  """
  Main function to get user input and print the Fibonacci series.
  """
  while True:
    try:
      # Get the number of terms from user input
      num_terms_str = input("Enter the number of terms for the Fibonacci series: ")
      num_terms = int(num_terms_str)
      # Check if the input is a non-negative integer
      if num_terms < 0:
        print("Please enter a non-negative integer.")
      else:
        break # Exit the loop if input is valid
    except ValueError:
      print("Invalid input. Please enter an integer.")

  # Generate the Fibonacci series
  fibonacci_result = generate_fibonacci_series(num_terms)

  # Print the Fibonacci series
  print(f"The Fibonacci series up to {num_terms} terms is:")
  print(fibonacci_result)

# Execute the main function
if __name__ == "__main__":
  main()
```

EXPLANATION:

1. `generate_fibonacci_series(n_terms)`returns a list of the first `n_terms` in the Fibonacci sequence, handling special cases for 0, 1, or 2 terms.
2. It starts with `[0, 1]` and keeps adding the sum of the last two numbers until the list reaches the desired length.
3. main() repeatedly prompts the user for a non-negative integer and handles invalid input.
4. Once valid input is received, it calls `generate_fibonacci_series()` and prints the result.
5. The `main()` function only runs if the script is executed directly (`if __name__ == "__main__"`).

```python
"""
Simple Calculator Module

This module provides basic arithmetic operations: addition, subtraction,
multiplication, and division. Each function takes two numbers as input
and returns the result of the operation.
"""

def add(a, b):
  """
  Adds two numbers.

  Args:
    a: The first number.
    b: The second number.

  Returns:
    The sum of the two numbers.
  """
  return a + b

def subtract(a, b):
  """
  Subtracts the second number from the first.

  Args:
    a: The first number.
    b: The second number.

  Returns:
    The difference between the two numbers.
  """
  return a - b

def multiply(a, b):
  """
  Multiplies two numbers.

  Args:
    a: The first number.
    b: The second number.

  Returns:
    The product of the two numbers.
  """
  return a * b

def divide(a, b):
  """
  Divides the first number by the second.

  Args:
    a: The first number (dividend).
    b: The second number (divisor).

  Returns:
    The quotient of the division.

  Raises:
    ZeroDivisionError: If the divisor is zero.
  """
  if b == 0:
    raise ZeroDivisionError("Cannot divide by zero.")
  return a / b

# Example usage:
if __name__ == "__main__":
  num1 = 10
  num2 = 5

  print(f"{num1} + {num2} = {add(num1, num2)}")
  print(f"{num1} - {num2} = {subtract(num1, num2)}")
  print(f"{num1} * {num2} = {multiply(num1, num2)}")

  try:
    print(f"{num1} / {num2} = {divide(num1, num2)}")
    print(f"{num1} / 0 = {divide(num1, 0)}")
  except ZeroDivisionError as e:
    print(f"Error: {e}")
```
